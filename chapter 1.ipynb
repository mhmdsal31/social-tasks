```python
import networkx as nx
%matplotlib inline
G = nx.Graph()
G.add_node('a')
nodes_to_add = ['b', 'c', 'd','e','f']
G.add_nodes_from(nodes_to_add)
G.add_edge('a', 'b')
edges_to_add = [('a', 'c'), ('b', 'c'), ('c', 'd'),('a','f'),('f','e')]
G.add_edges_from(edges_to_add)
nx.draw(G, with_labels=True)
```

```python
nx.draw(G,
        with_labels=True,
        node_color='green',
        node_size=1600,
        font_color='white',
        font_size=16,
        )

G.nodes()
G.edges()
for edge in G.edges:
    print(edge)
```

```python
for neighbor in G.neighbors('a'):
    print(neighbor)

len(list(G.neighbors('a')))

items = ['alex', 'cr7', 'mane']
[item.upper() for item in items]

g = (len(item) for item in items)
list(g)
```
